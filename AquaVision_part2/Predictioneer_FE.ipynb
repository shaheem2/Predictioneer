```python
#  Import some data manipulation and plotting packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

#ignore warnings
import warnings
warnings.filterwarnings("ignore")
```

```python
aquifer_auser = pd.read_csv("aquifer_auser_preprocessed.csv")
aquifer_auser.head()
```

```python
# create a new dataframe for aquifer auser daily average of all columns
aquifer_auser_daily_avg = pd.DataFrame()
aquifer_auser_daily_avg['Date'] = aquifer_auser['Date']
```

```python
# take mean of all rainfall features on daily date rows
aquifer_auser_daily_avg['Mean_Rainfall'] = aquifer_auser[['Rainfall_Gallicano', 'Rainfall_Pontetetto', 'Rainfall_Monte_Serra',
       'Rainfall_Orentano', 'Rainfall_Borgo_a_Mozzano', 'Rainfall_Piaggione',
       'Rainfall_Calavorno', 'Rainfall_Croce_Arcana',
       'Rainfall_Tereglio_Coreglia_Antelminelli',
       'Rainfall_Fabbriche_di_Vallico']].mean(axis=1)
```

```python
# take mean of all depth features on daily date rows
aquifer_auser_daily_avg['Depth_to_Groundwater_LT2'] = aquifer_auser['Depth_to_Groundwater_LT2']
aquifer_auser_daily_avg['Depth_to_Groundwater_SAL'] = aquifer_auser['Depth_to_Groundwater_SAL']
aquifer_auser_daily_avg['Depth_to_Groundwater_PAG'] = aquifer_auser['Depth_to_Groundwater_PAG']
aquifer_auser_daily_avg['Depth_to_Groundwater_CoS'] = aquifer_auser['Depth_to_Groundwater_CoS']

```

```python
# take mean of all temperature features on daily date rows
aquifer_auser_daily_avg['Mean_Temp'] = aquifer_auser[['Temperature_Orentano',
       'Temperature_Monte_Serra', 'Temperature_Ponte_a_Moriano',
       'Temperature_Lucca_Orto_Botanico']].mean(axis=1)
```

```python
# take mean of all volume features on daily date rows
aquifer_auser_daily_avg['Mean_Volume'] = aquifer_auser[['Volume_POL', 'Volume_CC1',
       'Volume_CC2', 'Volume_CSA', 'Volume_CSAL']].mean(axis=1)
```

```python
# take mean of all hydrometry features on daily date rows
aquifer_auser_daily_avg['Mean_Hydrometry'] = aquifer_auser[['Hydrometry_Monte_S_Quirico']].mean(axis=1)
```

```python
# take mean of all depth features on daily date rows
aquifer_auser_daily_avg['Mean_Depth'] = aquifer_auser[['Depth_to_Groundwater_LT2',
       'Depth_to_Groundwater_SAL', 'Depth_to_Groundwater_PAG',
       'Depth_to_Groundwater_CoS']].mean(axis=1)
```

```python
# create a new column to store the Aquifer name
aquifer_auser_daily_avg['Source'] = 'Auser'
```

```python
aquifer_auser_daily_avg
```

```python
# Convert to date
aquifer_auser_daily_avg['Date'] = pd.to_datetime(aquifer_auser_daily_avg['Date']).dt.date
```

```python
aquifer_auser_daily_avg
```

```python
aquifer_auser_daily_avg.info()
```

```python
# dataframe for shifted values of depth feature
shifted_depth = pd.DataFrame()
```

```python
#shift the depth values one day at a time
for i in range(1,32):
    shifted_depth['Shifted_Depth_'+str(i)] = aquifer_auser_daily_avg['Mean_Depth'].shift(-i)
```

```python
shifted_depth
```

```python
# fill the gap after shifted values with last value itself
shifted_depth = shifted_depth.ffill()
shifted_depth
```

```python
# take rainfall and temp features
X = aquifer_auser_daily_avg[['Mean_Rainfall','Mean_Temp']]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import median_absolute_error
```

```python
# create a random forest model to find the mae between predicted depth and shifted depth
mae_errors = []
for i in range(1,32):
    Y = shifted_depth['Shifted_Depth_'+str(i)]
    X_train, X_test, y_train, y_test = train_test_split(X,Y,test_size=0.2,random_state=12)
    rf = RandomForestRegressor()
    model = rf.fit(X_train, y_train)
    yhat = model.predict(X_test)
    mae = median_absolute_error(y_test,yhat)
    mae_errors.append(mae)
```

```python
# Plot the MAE on depth due to rainfall and temp
plt.title("Median Absolute Error on Depth due to Rain and Temp ")
plt.ylabel("MAE")
plt.xlabel('Day')
plt.plot(mae_errors)
```
